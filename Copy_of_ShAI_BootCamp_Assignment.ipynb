# Data Cleaning

```python
df['BasePay'].fillna(0, inplace=True)
df['BasePay'].isnull().sum()

#we replace the null values in the BasePay column with 0 because there are other types of pays,
#as we found earlier the number of null values in the basepay column is more than total pay
#which means 606 people were not getting a base pay but getting other type of payments.
```

```python
df['Benefits'].fillna(0, inplace=True)
df['BasePay'].isnull().sum()

#we do a similar approach to the Benefits column as well.
```

```python
df['OvertimePay'].fillna(0, inplace=True)
df['OtherPay'].fillna(0, inplace=True)
df['TotalPayBenefits'].fillna(0, inplace=True)
df['TotalPay'].fillna(0, inplace=True)
df.isnull().sum()

#I saw that it was best to fill with 0's for those columns as well
```

```python
df['Notes'].fillna('NaN', inplace=True)
df['Status'].fillna('NaN', inplace=True)
df.isnull().sum()
# Filling the missing values in columns of type string with NaN
```

# **Simple Correlation Analysis**

```python
correlation_coefficient = df['TotalPay'].corr(df['BasePay'])
print(f"Correlation between Salary and BasePay: {correlation_coefficient:.2f}")
#The correlation coefficient ranges from -1 to 1
#Positive value indicates a positive correlation (salary increases with Base Pay).
#Negative value indicates a negative correlation (salary decreases with Base Pay).
```

```python
df.plot.scatter(x='BasePay', y='TotalPay', title='Correlation between Salary and Base Pay')
plt.xlabel('BasePay')
plt.ylabel('TotalPay')
plt.annotate(f"Correlation: {correlation_coefficient}", xy=(0.8, 0.8))
plt.show()
```
